```python
class LinearRegression(torch.nn.Module):
  def __init__(self):
    super(LinearRegression, self).__init__()
    self.layers = nn.Sequential(
        nn.BatchNorm1d(x_train.shape[1]),
        nn.Linear(x_train.shape[1], 1)
    )
    self.norm = nn.BatchNorm1d(x_train.shape[1])
    self.linear = nn.Linear(x_train.shape[1], 1)
  def forward(self, x):
    self.x = x.float()
    # self.x_norm = self.norm(self.x)
    self.y_pred = self.linear(self.x)
    
    return self.y_pred
```

## Training

```python
lr = 1e-2
max_epoch = 1000
batch_size = x_train.shape[0]
batch_num = math.ceil(x_train.shape[0] / batch_size)
```

```python
loss_func = nn.MSELoss().to(device)
model = LinearRegression()
model = model.to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=lr)
```

```python
model.train()

for epoch in range(max_epoch):
  e_loss = 0

  for x_batch, y_batch in loader:
    optimizer.zero_grad()

    y_pred = model(x_batch)
    loss = loss_func(y_pred.double(), y_batch.double()).to(device)
    e_loss += loss.item()

    loss.backward()
    optimizer.step()

  e_loss = e_loss / batch_num
  print('epoch: {}, loss: {}'.format(epoch, e_loss))
  # input()
```
